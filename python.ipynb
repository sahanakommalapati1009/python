## Using return

Let’s see some examples that use a return statement. Return allows a function to “return” a result that can then be stored as a variable, or used in whatever manner a user wants.

```python
def add_num(num1,num2):
    return num1+num2
result = add_num(4,5)
```

```python
def func_with_print():
    print("Hello")

def func_with_return():
    return "Hello"

# Call both and store results
a = func_with_print()
b = func_with_return()

print("a =", a)   # Will print: a = None
print("b =", b)   # Will print: b = Hello
```

```python
def is_prime(num):
    '''
    Naive method of checking for primes.
    '''
    for n in range(2,num):
        if num % n == 0:
            print('not prime')
            break
    else: # If never mod zero, then prime
        print('prime')
```

```python
is_prime(16)
```

## Generators
- A generator is a special type of iterator that is written like a function but uses the yield keyword instead of return.
- It pauses when it yields, and resumes from where it left off.

The main difference between yield and return are
- return ends the function and sends back a value.
- yield pauses the function and remembers its state, allowing it to resume later.

```python
# Generator function for the cube of numbers (power of 3)
def gencubes(n):
    for num in range(n):
        yield num**3
```

It only creates a generator object.

That generator object:

remembers the function (gencubes)

remembers the current position (num)

knows how to produce the next cube when asked

```python
gencubes(10)
```

A generator does not give values automatically. It only prepares the logic.
To actually get the values, we must iterate over it.

```python
for x in gencubes(9):
    print(x)
```

#### Example: Fibonacci generator

```python
def genfibonacci(n):
    a=1
    b=1
    for i in range(n):
        yield a
        a, b = b, a+b
```

What this function does:

- It does not build a full Fibonacci list.

- It creates a generator that produces one Fibonacci number at a time.

```python
genfibonacci(10)
```

It only creates the generator object.

The generator will start producing values only when you iterate over it.

```python
for num in genfibonacci(10):
    print(num)
```

The for loop:
- Automatically calls next() on the generator
- Gets one value at a time
- Stops when the sequence ends

Why is this better than using a list?
- If we create a list of 1 million Fibonacci numbers, It would use a lot of RAM.
- But a generator, Stores only two numbers at a time. Uses almost zero memory
- This is why generators are very efficient.


## Maps()
- The map() is a function that takes in two arguments:
1. A function
2. A sequence iterable
- It is In the form of map(function, sequence)
- The first argument is the name of a function and the second a sequence (e.g. a list).
- map() applies the function to all the elements of the sequence.
- It returns a new list with the elements changed by the function.
- Short and clean

```python
##  Example: convert Fahrenhiet to celcius

def fahrenheit(T):
    return ((float(9)/5)*T + 32)

def celsius(T):
    return (float(5)/9)*(T-32)
```

```python
temp = [0, 22.5, 40, 100]
```

```python
for i in temp:
    celsius = (float(5)/9)*(i-32)
    print(celsius)
```

```python
F_temps = list(map(celsius, temp))
F_temps
```

```python
# convert back to Fahrenheit
C_temps = list(map(fahrenheit, F_temps))
C_temps
```

## Object Oriented Programming

- OOP is a programming paradigm based on objects and classes.
- It helps organize code by grouping data and behavior together.
- Improves code reusability, scalability, and maintainability.
- Python supports OOP features like class, object, inheritance, and polymorphism.

## Creating a Class

- A class is a blueprint used to create objects.
- Classes are defined using the class keyword.
- Class names should follow PascalCase (start with a capital letter).
- A class can contain attributes (data) and methods (functions).

```python
class Person:
    def __init__(self, name, surname, age):
        self.name = name
        self.surname = surname
        self.age = age
```

- Person is the class name and represents a general concept.
- __init__() is a constructor that initializes object data.
- name, surname, and age are attributes of the class.
- self refers to the current object instance.

```python
p1 = Person("Sahana", "Kommalapati", 24)
print(p1.name, p1.surname, p1.age)
```

- p1 is an instance of the Person class.
- Values are passed to the constructor during object creation.
- Attributes are accessed using dot (.) notation.
- Each object has its own data in memory.

## Methods

- Methods are functions defined inside a class
- They describe the behavior of an object
- self allows access to object attributes
- Methods can perform actions or return values

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name}")

    def is_adult(self):
        return self.age >= 18

p1 = Person("Sahana", 24)
p1.greet()
print(p1.is_adult())
```

- greet() and is_adult() are methods of the Person class.
- self.name and self.age access object attributes.
- greet() prints a message using object data.
- is_adult() returns a value based on object state.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Name: {self.name}, Age: {self.age}"

p1 = Person("Sahana", 24)
print(p1)
```

- __str__() is a special (magic) method in Python.
- It defines how an object is displayed as a string.
- It is automatically called by print() and str().
- Makes object output more readable and meaningful.

```python
# Without __str__()
class Person:
    pass

p2 = Person()
print(p2)
```

## Class Without __init__() Method

- It is possible to create a class without the __init__() method
- Such classes do not initialize object attributes automatically
- Objects created from these classes may not represent consistent data
- This is considered a bad practice in most cases

```python
class Person:
    def set_name(self, name):
        self.name = name
        
    def set_surname(self, surname):
        self.surname = surname
        
    def set_year_of_birth(self, year_of_birth):
        self.year_of_birth = year_of_birth
        
    def age(self, current_year):
        return current_year - self.year_of_birth

    def __str__(self):
        return "%s %s was born in %d." % (self.name, self.surname, self.year_of_birth)

# Creating and using an object
p1 = Person()
p1.set_name("Sahana")
p1.set_surname("Kommalapati")
p1.set_year_of_birth(1999)

print(p1)               # Deepak Nallapaneni was born in 1999.
print(p1.age(2025))     # 26
```

### Inheritance
- Inheritance means: one class can take (reuse) features of another class.

- A parent has some qualities.
